## To force rendering in pyodide

Add the following to quarto



```{pyodide}
#| echo: false
#| warning: false
#| message: false

import matplotlib.pyplot as plt
from pandas.plotting._core import PlotAccessor

_old_call = PlotAccessor.__call__

def _patched_call(self, *args, **kwargs):
    ax = _old_call(self, *args, **kwargs)
    plt.show()
    return ax

PlotAccessor.__call__ = _patched_call
```
